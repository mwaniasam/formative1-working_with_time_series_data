```python
import requests
import pandas as pd
import os
```

```python
response = requests.get("https://unantlered-ryleigh-phlegmatically.ngrok-free.dev/mongo/records")
data = response.json()

print(len(data))
print(data[0])
```

```python
def flatten_record(record):
    flat = {}

    # Timestamp
    flat['time'] = record['timestamp']

    # Load
    flat['total load actual'] = record['load']['actual']
    flat['total load forecast'] = record['load']['forecast']

    # Price
    flat['price actual'] = record['price']['actual']
    flat['price day ahead'] = record['price']['day_ahead']

    # Generation
    flat['generation biomass'] = record['generation']['biomass']['mw']
    flat['generation fossil brown coal'] = record['generation']['fossil_brown_coal']['mw']
    flat['generation fossil coal-derived gas'] = record['generation']['fossil_coal_derived_gas']['mw']
    flat['generation fossil gas'] = record['generation']['fossil_gas']['mw']
    flat['generation fossil hard coal'] = record['generation']['fossil_hard_coal']['mw']
    flat['generation fossil oil'] = record['generation']['fossil_oil']['mw']
    flat['generation fossil oil shale'] = record['generation']['fossil_oil_shale']['mw']
    flat['generation fossil peat'] = record['generation']['fossil_peat']['mw']
    flat['generation geothermal'] = record['generation']['geothermal']['mw']
    flat['generation hydro pumped storage consumption'] = record['generation']['hydro_pumped_storage_cons']['mw']
    flat['generation hydro run-of-river and poundage'] = record['generation']['hydro_run_of_river']['mw']
    flat['generation hydro water reservoir'] = record['generation']['hydro_water_reservoir']['mw']
    flat['generation marine'] = record['generation']['marine']['mw']
    flat['generation nuclear'] = record['generation']['nuclear']['mw']
    flat['generation other'] = record['generation']['other']['mw']
    flat['generation other renewables'] = record['generation']['other_renewable']['mw']
    flat['generation solar'] = record['generation']['solar']['mw']
    flat['generation solar forecast'] = record['generation']['solar']['forecast_mw']
    flat['generation waste'] = record['generation']['waste']['mw']
    flat['generation wind offshore'] = record['generation']['wind_offshore']['mw']
    flat['generation wind onshore'] = record['generation']['wind_onshore']['mw']
    flat['generation wind onshore forecast'] = record['generation']['wind_onshore']['forecast_mw']

    return flat
```

```python
flat_record = flatten_record(data[0])
print(flat_record)
```

```python
flat_records = [flatten_record(record) for record in data]
df = pd.DataFrame(flat_records)
print(df.shape)
print(df.head())
```

```python
df['time'] = pd.to_datetime(df['time'])
df.set_index('time', inplace=True)
# Time features
df['hour']        = df.index.hour
df['day_of_week'] = df.index.dayofweek
df['month']       = df.index.month
df['quarter']     = df.index.quarter
df['is_weekend']  = (df.index.dayofweek >= 5).astype(int)

# Lag features
# Price lags: 1h ago, 24h ago (same hour yesterday), 168h ago (same hour last week)
for lag in [1, 24, 168]:
    df[f'price_lag_{lag}h'] = df['price actual'].shift(lag)

# Load lags
for lag in [1, 24]:
    df[f'load_lag_{lag}h'] = df['total load actual'].shift(lag)

# Moving average features
# shift(1) ensures no data leakage — we only use past values
df['price_ma_24h']  = df['price actual'].shift(1).rolling(24).mean()
df['price_ma_168h'] = df['price actual'].shift(1).rolling(168).mean()
df['price_std_24h'] = df['price actual'].shift(1).rolling(24).std()
df['load_ma_24h']   = df['total load actual'].shift(1).rolling(24).mean()

# Drop NaN rows from lag/rolling
df.dropna(inplace=True)
print(f'Final shape after feature engineering: {df.shape}')
```

```python
import joblib
import tensorflow as tf

# Load the model
model = tf.keras.models.load_model("/content/model_lstm_deep.keras")

# Load the scalers
scaler_X = joblib.load("/content/scaler_minmax_X.pkl")
scaler_y = joblib.load("/content/scaler_minmax_y.pkl")

print("Model and scalers loaded successfully!")
```

```python
import json

with open('feature_columns.json', 'r') as f:
    feature_columns = json.load(f)

print(len(feature_columns))
print(feature_columns)
```

```python
# Check which feature columns are missing from your df
missing = [col for col in feature_columns if col not in df.columns]
print("Missing columns:", missing)

# Check which columns in df are not in feature_columns
extra = [col for col in df.columns if col not in feature_columns]
print("Extra columns:", extra)
```

```python
df.rename(columns={
    'generation fossil brown coal': 'generation fossil brown coal/lignite',
    'generation other renewables': 'generation other renewable',
    'generation solar forecast': 'forecast solar day ahead',
    'generation wind onshore forecast': 'forecast wind onshore day ahead'
}, inplace=True)
```

```python
missing = [col for col in feature_columns if col not in df.columns]
print("Missing columns:", missing)

extra = [col for col in df.columns if col not in feature_columns]
print("Extra columns:", extra)
```

```python
X = df[feature_columns].values
X_scaled = scaler_X.transform(X)
print(X_scaled.shape)
```

```python
LOOKBACK = 24
X_input = X_scaled[-LOOKBACK:]
X_input = X_input.reshape(1, LOOKBACK, X_scaled.shape[1])

print(X_input.shape)
```

```python
# Make prediction
prediction_scaled = model.predict(X_input)

# Inverse transform to get actual price in €/MWh
prediction = scaler_y.inverse_transform(prediction_scaled)

print(f"Predicted electricity price: {prediction[0][0]:.2f} €/MWh")
```
